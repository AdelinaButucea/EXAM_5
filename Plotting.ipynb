```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sn
```

```python
data = pd.read_csv('data_ready.csv')
```

**CHECK FOR CORRELATIONS**

```python
fig, ax = plt.subplots(9, 2, figsize=(15, 30), sharex=True)
for index, column in enumerate(data.columns[:9]):
    c = abs(data.corr()['Price'][column])
    c_label = data.corr()['Price'][column]
    ax[index][0].scatter(data.Price, data[column], color = (0.8 * (1-c), 0.8 * c, 0.3, 0.5), label=c_label)
    ax[index][0].set_ylabel(column)
    ax[index][0].legend()

for index, column in enumerate(data.columns[9:]):
    c = abs(data.corr()['Price'][column])
    c_label = data.corr()['Price'][column]
    ax[index][1].scatter(data.Price, data[column], color = (0.8 * (1-c), 0.8 * c, 0.3, 0.5), label=c_label)
    ax[index][1].set_ylabel(column)
    ax[index][1].legend()
```

Each color represents the level of correlation.

**CHECK FOR OUTLIERS**

```python
fig, ax = plt.subplots(9, 2, figsize=(15, 30), sharex=True)
for index, column in enumerate(data.columns[:9]):
    ax[index][0].boxplot(data[column], whiskerprops={'color':'r'}, flierprops = {'color' : (0.3, 0.2, 0.1 * index)}, 
                         patch_artist=True, boxprops={'facecolor':((np.sqrt(index/18), index/18 ** 2, index/18 ** 3, 0.3))})

    ax[index][0].grid(linestyle='--')
    ax[index][0].set_ylabel(column, fontdict={'size':15})


for index, column in enumerate(data.columns[9:]):
    ax[index][1].grid(linestyle='--')
    ax[index][1].boxplot(data[column], patch_artist=True, boxprops={'facecolor':((np.sqrt(index/18), index/18 ** 2, index/18 ** 3, 0.3))})
    ax[index][1].set_ylabel(column, fontdict={'size':15})
```
